```python
driver = webdriver.Chrome("chromedriver.exe")
car_urls = []
for page in range(0,len(loc_urls)):
    driver.get(loc_urls[page])
    for button in range(0,15):
        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
            time.sleep(5)
            driver.find_element_by_xpath("//li[@class='_20FqS']/div/button").click()
            time.sleep(5)
        except:
            pass
    time.sleep(5)
    elem_urls = driver.find_elements_by_tag_name("a")
    for element in elem_urls:
        if '/item/' in element.get_attribute("href"):
            car_urls.append(element.get_attribute("href"))
    print(f"Scrapped page: {page}, URLs Collected:{len(car_urls)}")
```
